```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("flights")
```

```python
# DATA PIPELINE SIMULATION

df['anomaly_flag'] = df['passengers'].apply(lambda x: 1 if x > df['passengers'].mean() + 2*df['passengers'].std() else 0)
```

```python
# PIPELINE KPI

print("Total Rows:", len(df))
print("Anomalies:", df['anomaly_flag'].sum())
```

```python
# VISUALIZATION

plt.figure()
df.groupby('year')['passengers'].sum().plot()
plt.title("Passenger Trend Over Years")
plt.show()
```

```python
plt.figure()
df[df['anomaly_flag']==1]['passengers'].plot(kind='bar')
plt.title("Detected Anomalies")
plt.show()
```
